```python
def length_of_string(s):
    l = 0
    for i in s:
        l += 1
    return l

p = "Hello"
print(length_of_string(p))
```

```python
def dict_to_list(dic,a):
    for key,value in dic.items():
        if (type(value) is dict):
            dict_to_list(value,a)
        else:
            a.append(value)

d = {"a":1,"b":{"c":{"d":2}},"e":4}
li = []

dict_to_list(d,li)
print(li)
```

```python
def concat_list(*args):
    A = []
    for L in args:
        for item in L:
            A.append(item)
    return A


lista = [1,2]
listb = [3,4]

concat_list_list = concat_list(lista,listb)

print(concat_list_list)
```

```python
def find_index(listc,value):
    A = []
    for index,item in enumerate(listc):
        if item == value:
            A.append(index)
    return A

lista = [1,2,2,6,7,2]

indexs = find_index(lista,2)

print(indexs)
```
